```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
np.random.seed(42)

mu0, sigma0, n0 = 0.4, 0.05, 40
mu1, sigma1, n1 = 0.6, 0.05, 60
X0, X1 = norm.rvs(mu0, sigma0, n0), norm.rvs(mu1, sigma1, n1)
```

```python
plt.scatter(X0, np.zeros_like(X0), marker='.', c='r', label='Y=0')
plt.scatter(X1, np.zeros_like(X1), marker='x', c='b', label='Y=1')
plt.legend()
plt.ylim([-0.05, 1])
plt.xlabel('Umidade relativa média do ar')
```

```python
xs0 = np.linspace(0.25, 0.75)
ys0 = norm.pdf(xs0, mu0, sigma0)

ys1 = norm.pdf(xs0, mu1, sigma1)

plt.scatter(X0, np.zeros_like(X0), marker='.', c='r', label='Y=0')
plt.scatter(X1, np.zeros_like(X1), marker='x', c='b', label='Y=1')

plt.plot(xs0, ys0, c='r')
plt.plot(xs0, ys1, c='b')
plt.legend()
# plt.ylim([-0.05, 1])
plt.xlabel('Umidade relativa média do ar')
```

```python
xs0 = np.linspace(0.25, 0.75)
ys0 = norm.pdf(xs0, mu0, sigma0)

ys1 = norm.pdf(xs0, mu1, sigma1)

probs = ys1 * 0.6 / (ys0 * 0.4 + ys1 * 0.6)

plt.scatter(X0, np.zeros_like(X0), marker='.', c='r', label='Y=0')
plt.scatter(X1, np.zeros_like(X1), marker='x', c='b', label='Y=1')

plt.plot(xs0, ys0, c='r')
plt.plot(xs0, ys1, c='b')
plt.plot(xs0, probs, c='k', label='P(Y=1|x)')
plt.legend()
# plt.ylim([-0.05, 1])
plt.xlabel('Umidade relativa média do ar')
```
